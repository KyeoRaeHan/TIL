```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic
df.head()
```

```python
df_af = titanic.loc[:,['age','fare']]
df_af
```

```python
# Q.10을 더하는 함수 add_10 과 두 객체의 합 add_two_obj 함수를 정의한 후 
# 결과값이 각각 20과 20을 출력하세요.¶
```

```python
def add_10(n):
    return n + 10

def add_two_obj(a,b):
    return a + b

print(add_10(10))
print(add_two_obj(10,10))
```

```python
# Q. 10을 곱하는 사용자 함수와 두 객체의 나눗셈을 수행하는 사용자 함수를 정의한 후 
# 결과값이 각각 200과 10을 출력하세요.
```

```python
def mul_10(n):
    return n * 10

def div_two_obj(a,b):
    return a / b

print(mul_10(10))
print(div_two_obj(10,10))
```

```python
df_age = df_af['age']
df_age

sr1 = df_age.apply(add_10)
```

```python
# Q. 시리즈 객체 + 숫자를 연산하는 방식으로 df_af.age에 
# 10을 더한 후 sr2에 저장하고 출력하세요.
```

```python
def add_two_obj(a,b):
    return a + b

sr2 = df_af.apply(add_two_obj, b=10)
sr2.head()
```

```python
# Q. 람다 함수를 활용해서 위와 같은 연산을 수행하고 결과를 sr3로 출력하세요.
```

```python
sr3 = df_af.age.apply(lambda x: add_10(x))
sr3.head()
```

```python
# Q. df_af의 'age', 'fare' 칼럼에 10을 더하는 사용자 함수를 정의하고 
# applymap()으로 각 사용자 함수를 매핑한 후 출력하세요.
```

```python
df_map = df_af.applymap(add_10)
df_map.head()
```

```python
# Q. df_map의 [0,0]을 NaN으로 변경한 후 
# missing_value 함수를 적용해서 불린 시리즈를 반환하세요.¶
```

```python
import numpy as np

def missing_value(series):
    return series.isnull()

df_map.iloc[0,0] = np.nan
print(df_map)
result = df_map.apply(missing_value,axis=0)
result.head()
```

```python
# Q. titanic 데이터셋을 seaborn에서 로딩한 후 블린 인덱싱을 이용하여 아래 사항들을 수행하세요.
# 나이가 10대(10~19세)인 승객만 따로 선택
# 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()

mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic.loc[mask1, :]

df_teenage.head()
```

```python
mask2 = (titanic.age < 10 ) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2, :]
df_female_under10.head()
```

```python
mask3 = (titanic.age < 10 ) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```

```python
# Q. 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 alone 열만 
# 선택한 후 값을 True로 변경하세요
```

```python
df_m3 = titanic.loc[(titanic.age < 10) | (titanic.age >=60),'alone']
print(df_m3)
titanic.loc[(titanic.age < 10) | (titanic.age >=60),'alone'] = True
df_m5 = titanic.loc[(titanic.age < 10) | (titanic.age >=60),'alone']
df_m5
```
