```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
titanic = pd.read_csv('C:/caba2012/workspace/Caba_dataprocessing/dataset/titanic3.csv')
titanic.head()
```

```python
titanic['sex'].isnull().sum()
```

```python
titanic['embarked'].fillna('N', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

def enconde_feature(dataDF):
    features = ['sex', 'embarked']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF
```

```python
titanic_encode = enconde_feature(titanic)
titanic_encode.head()
```
