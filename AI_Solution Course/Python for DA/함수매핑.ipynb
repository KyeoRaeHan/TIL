```python
# fifa_stat
# Q2. 포지션을 의미하는 Position변수를 아래 표를 참고하여 “Forward”, “Midfielder”,“Defender”, “GoalKeeper”로 재범주화하고, 
# 변환하여 Position_Class 라는 변수를 생성하고 저장하시오.

# Forward : ['LS','ST','RS','LW','LF','CF','RF','RW']
# Midfielder : ['LAM','CAM','RAM','LM','LCM','CM','RCM','RM']
# Defender : ['LBW','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
# GoalKeeper
```

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/caba2012/workspace/Caba_stat/dataset-20210111T020713Z-001/dataset/FIFA.csv', engine='python')
```

```python
# Forward : ['LS','ST','RS','LW','LF','CF','RF','RW']
# Midfielder : ['LAM','CAM','RAM','LM','LCM','CM','RCM','RM']
# Defender : ['LBW','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
# GoalKeeper
```

```python
def cat_position(str):
    if str in ['LS','ST','RS','LW','LF','CF','RF','RW']:
        return 'Forward'
    
    elif str in ['LAM','CAM','RAM','LM','LCM','CM','RCM','RM']:
        return 'Midfielder'
    
    elif str in ['LBW','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']:
        return 'Defender'
    
    else:
        return 'Goalkeeper'
```

```python
a = data['Position'].apply(cat_position)
```

```python
data['Position_Class'] = a
```

```python
data.head()
```

```python
data['Position'] = data['Position_Class']
```

```python
#position 컬럼 삭제
data.drop(['Position'], axis=1)
```
