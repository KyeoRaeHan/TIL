```python
[선형대수]

스칼라
1차원 벡터를 스칼라라고 한다. 스칼라는 크기만 있고 방향성이 존재하지 않습니다. 
스칼라는 움직일 수 있는 방향이 하나뿐이므로 그 방향은 중요하지 않으며, 
크기만 고려하면 됩니다. 그 예로 키 혹은 무게 등이 있음

벡터
연속 또는 이산 숫자의 배열을 벡터라 하고, 벡터로 구성된 공간을 벡터 공간이라고 한다. 
벡터 공간 차원은 유한하거나 무한하지만, 
대부분의 머신 러닝 또는 데이터 과학에서는 고정 길이 벡터를 처리. 
머신러닝에서는 다차원 데이터를 처리하기 때문에 벡터가 아주 중요

Data들을 표현하는 '특징(feature)'을 선형대수학의 기본 단위인 vector로 나타낼 수 있다.
예를 들어, 집의 면적, 침식의 수, 욕실의 수, 지역의 인구 밀도에 기초해 
주택 가격을 예측하려 한다고 가정하면, 
이러한 모든 특징을 이용해 주택 가격 예측 문제에 대한 입력 특성 벡터를 만들 수 있다.
벡터의 모임인 행렬을 통해 방정식 연산을 훨씬 빠르게 할 수 있다.

텐서
텐서는 숫자의 다차원 배열입니다. 사실, 벡터와 행렬은 1D 및 2D 텐서로 취급될 수 있습니다.
딥러닝에서 텐서는 주로 데이터 저장 및 처리에 사용됩니다. 
예를 들어 RGB의 이미지는 3차원 텐서에 저장되며, 
여기서 한 차원은 가로축, 다른 차원은 세로축, 그리고 마지막 차원은 세 가지 색 채널, 
즉 Red, Green, Blue를 저장합니다. 
또 다른 예로, 컨볼루션 신경망에서 미니 배치를 통해 이미지를 공급하는데 사용되는 4차원 텐서입니다. 
첫 번째 차원에는 일괄 처리의 이미지 번호가 있고, 두 번째 차원에는 색상 채널이 있으며, 
세 번째 및 네 번째 차원에는 가로 및 세로 방향의 픽셀 위치가 있습니다.
```

```python
# 벡터의 내적
a = np.arange(1,4)
b = np.arange(4,7)
#행렬의 곱연산
print(np.dot(a,b))

# 행렬의 곱
A = np.arange(1,5).reshape(2,2)
B = np.arange(5,9).reshape(2,2)
print(A)
print(B)
# matmul과 dot의 차이
# 3차원 행렬에서 다른 결과, 상수 곱은 dot만 가능
# dot은 내적곱에 사용, 행렬곱에는 matmul을 사용할 것을 권장
# https://cyber0946.tistory.com/64
print(np.matmul(A,B))
print(np.dot(A,B))
```

```python
# 완전연결계층에 의한 변환은 선형변환인데 
# 여기에 비선형 효과를 부여하는 것이 '활성화 함수'
# 비선형 활성화 함수를 이용함으로써 '신경망의 표현력'을 높일 수 있다. 
# 시그모이드 함수(이진분류에 주로 사용)는 임의의 실수를 입력받아 0~1사이의 실수 출력

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
#  함수에 하나의 값을 입력하면 exponential 값을 반환합니다.
a = np.exp(0) # e**0
b = np.exp(2) # e**2
print(a,b)
```

```python
# array는 리스트를 배열 형태로 바꿔준다
# 1X2X2X3 = 1 X 3
X = np.array([1.0, 0.5])
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])
A0 = np.dot(X, W1)
print(A0)
A1 = np.dot(X, W1) + B1
print(A1)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
Z1 = sigmoid(A1)
print(A1)
print(Z1)
```
